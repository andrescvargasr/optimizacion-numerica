# 1. Introducción ([Chong & Zak, 2001](https://eng.uok.ac.ir/mfathi/Courses/Advanced%20Eng%20Math/An%20Introduction%20to%20Optimization-%20E.%20Chong,%20S.%20Zak.pdf); [Bertsekas, 2008](http://venus.unive.it/~fasano/Nonlinear_Programming_IIed.pdf))

La optimización es importantísima para cualquier problema que involucra la toma de decisiones ya sea en ingeniería o en economía. La tarea de tomar decisiones implica elegir entre varias alternativas. Esta decisión es gobernada por nuestro deseo de tomar la "mejor" decisión. Las medidas de bondad de las alternativas es descrita por una función objetivo o índice de desempeño. La teoría y métodos de optimización conllevan a seleccionar la mejor alternativa respecto a una función objetivo.

El área de la optimización ha recibido mucha atención en los últimos años, ya que se ha visto estimulado por **"El rápido crecimiento de las tecnologías de la información, incluyendo el desarrollo y disponibilidad de software amigable, con procesadores de alta velocidad y/o paralelos, y las redes neuronales ,.artificiales"**.

## 1.2. Notación Básica ([Math Symbols](https://en.wikipedia.org/wiki/List_of_mathematical_symbols))

(Para comprender parte de las notaciones que se usarán en este curso)

Si $X$ es un conjunto

Y se tiene que:

$X \subset \mathbb{R}^{n}$ significa: "Cada elemento de $X$ es tambien un elemento de $\mathbb{R}^{n}$". Por otro lado, se puede afirmar que: "Los elementos del conjunto $X$ son números reales en $n$-dimensiones"

$x \in X \Rightarrow$ "$x$ es un elemento de $X$"

$f: \mathbb{R}^{n}\to \mathbb{R} \Rightarrow$ "$f$ es una función representada en el conjunto de números reales en $n$-dimensiones de donde se obtienen números reales en una sola dimensión" ó "$f$ es una función evaluada en un sistema real de $n$-dimensiones que retorna un valor real".

El símbolo $:=$ denota asignación aritmética, por ejemplo:

$x := y \Rightarrow$ $x$ queda definida como $y$

Finalmente, el símbolo $\triangleq$ significa "igual por definición"
